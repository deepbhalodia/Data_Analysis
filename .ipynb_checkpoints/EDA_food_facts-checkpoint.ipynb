```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns

# Make plots larger
plt.rcParams['figure.figsize'] = (15, 9)
```

### Since Data is too large (333824 rows & 162 columns) let's look at how much data is NULL or NaN in each column so that we can get rid of columns which are highly NULL

```python
df = pd.read_table("dataset/en.openfoodfacts.org.products.tsv")
plt.figure(figsize=(5, 20))
df.isnull().mean(axis=0).plot.barh()
plt.title("Proportion of NaNs in each column")
```

```python
# If majorty of data is missing than interpolation will be highly skewed so ,
# Let's drop the columns where 70% or more are NULL
# 70% of 333824 = 233676
df=df.dropna(axis=1,thresh=233676)
df.drop('states', axis=1,inplace=True)
df.drop('states_en', axis=1,inplace=True)
df.drop('states_tags', axis=1,inplace=True)
df.drop('url', axis=1,inplace=True)
df.drop('created_datetime', axis=1,inplace=True)
df.drop('creator', axis=1,inplace=True)
df.drop('created_t', axis=1,inplace=True)
df.drop('last_modified_t', axis=1,inplace=True)
df.drop('last_modified_datetime', axis=1,inplace=True)
df.drop('countries_tags', axis=1,inplace=True)
df.drop('countries', axis=1,inplace=True)
df.drop('brands', axis=1,inplace=True)
# 70% of 17 = 12
df.dropna(axis=0,how='all',inplace=True,thresh=12)
df.apply(lambda x: sum(x.isnull()),axis=0)
df = df[df.ingredients_text.notnull()]
df = df[df.product_name.notnull()]
df = df[df.countries_en.notnull()]
df = df[df.brands_tags.notnull()]
df = df[df.additives.notnull()]

# sugars_100g and proteins_100g have negative values which are obviously not legit so let's remove them 
df.loc[df['sugars_100g'] < 0, 'sugars_100g'] = 0
df.loc[df['proteins_100g'] < 0, 'proteins_100g'] = 0

# Similarly other columns have outliers, so Removing outliers
df = df[np.abs(df.energy_100g-df.energy_100g.mean())<=(3*df.energy_100g.std())]
df = df[np.abs(df.fat_100g-df.fat_100g.mean())<=(3*df.fat_100g.std())]
df = df[np.abs(df['saturated-fat_100g']-df['saturated-fat_100g'].mean())<=(3*df['saturated-fat_100g'].std())]
df = df[np.abs(df.carbohydrates_100g-df.carbohydrates_100g.mean())<=(3*df.carbohydrates_100g.std())]
df = df[np.abs(df.sugars_100g-df.sugars_100g.mean())<=(3*df.sugars_100g.std())]
df = df[np.abs(df.proteins_100g-df.proteins_100g.mean())<=(3*df.proteins_100g.std())]
df = df[np.abs(df.salt_100g-df.salt_100g.mean())<=(3*df.salt_100g.std())]
df = df[np.abs(df.sodium_100g-df.sodium_100g.mean())<=(3*df.sodium_100g.std())]


# Filling Mean values where NaN is present
df['energy_100g'].fillna(df['energy_100g'].mean(), inplace=True)
df['fat_100g'].fillna(df['fat_100g'].mean(), inplace=True)
df['saturated-fat_100g'].fillna(df['saturated-fat_100g'].mean(), inplace=True)
df['carbohydrates_100g'].fillna(df['carbohydrates_100g'].mean(), inplace=True)
df['sugars_100g'].fillna(df['sugars_100g'].mean(), inplace=True)
df['proteins_100g'].fillna(df['proteins_100g'].mean(), inplace=True)
df['salt_100g'].fillna(df['salt_100g'].mean(), inplace=True)
df['sodium_100g'].fillna(df['sodium_100g'].mean(), inplace=True)
```

```python
#data distribution
plt.title("Protein Data Distribution")
sns.distplot(df.proteins_100g,axlabel="Protein/100gm",color="c")
```

```python
def splitDataFrameList(df, target_column, separator):
    def splitListToRows(row, row_accumulator, target_column, separator):
        split_row = row[target_column].split(separator)
        for s in split_row:
            new_row = row.to_dict()
            new_row[target_column] = s
            row_accumulator.append(new_row)
    new_rows = []
    df.apply(splitListToRows,axis=1,args = (new_rows,target_column,separator))
    new_df = pd.DataFrame(new_rows)
    return new_df
```

```python
food_ingredients = splitDataFrameList(df, "ingredients_text", ",")
ingredients = food_ingredients["ingredients_text"].value_counts()
plt.title("Top ingredients in food products")
ingredients[:20][::-1].plot.barh()
```
